## 3. Prueba de ingesta batch – Libros (Open Library)

```python
subject = 'scifi'
url = f'https://openlibrary.org/subjects/{subject}.json?limit=10'
response = requests.get(url)
data = response.json()
df_batch = pd.json_normalize(data['works'])
df_batch = df_batch[['title', 'key', 'first_publish_year']]
print(f'Batch obtenido: {len(df_batch)} libros.')
```
